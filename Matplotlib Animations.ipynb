```python
from pylab import *

delta_t = 0.01
#t = [i*delta_t for i in range(10)], HERE your interval is too short to see the oscillation. Try 1000
t = [i*delta_t for i in range(1000)]
a = []
v = []
x = []
x0 = 0
v0 = 0
#k = 1
#m = 5 Python thinks these are integers. Try:
k=1.0
m=1.0
g = 9.8

def acc(x):

    return -g -(k/m)*x

#print (t)

for i in range(len(t)):

    if t[i] == 0:

        x.append(x0)
        v.append(v0)
        a.append(acc(x0))

    else:
        v.append(v[i-1] + a[i-1]*delta_t)
        x.append(x[i-1] + v[i]*delta_t)
        a.append(acc(x[i]))

plot(t,x)
plot(t,v)

#To better monitor what is going on, you can also plot acceleration

plot(t,a,label="acceleration")
plot(t,v,label="velocity")
plot(t,x,label="position")

#The command below allows you to see legend, handy for multi-curve plots

legend()

#Let's also add some axis labels to make the plot more informative

xlabel("time[sec]")
ylabel("a(t),v(t),x(t)")
show()
```
