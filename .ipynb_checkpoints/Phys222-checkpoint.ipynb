```python
%pylab inline
from pylab import *
import numpy as np
#ion() #turning on the ‘‘interactive mode’’ in Python.

Lambda = 50
k = 2*pi/Lambda
omega = 3

dk = .1
domega = dk*50/pi

#See one full period
x = arange(0,Lambda*5,Lambda/200)

def first_wave(x,t):
    return cos(k*x-omega*t) #with input x and t, returns a value.

def second_wave(x,t):
    return cos((k+dk)*x-(omega+domega)*t) #with input x and t, returns a value.

traveling_wave, = plot(x, first_wave(x,0) + second_wave(x,0))

text(0.35*max(x), 1.2, 'wave (phase) velocity=')
text(0.7*max(x), 1.2, omega/k)

#Set the limits of the plot
xlim((0, max(x)))
ylim((-4,4))
xlabel('Time')
ylabel('$\psi$')

t = 0
while t < 50: # show the wave propagate for "some time"

    traveling_wave.set_ydata(first_wave(x,t) + second_wave(x,t))
    #wave2.set_ydata(second_wave(x,t))

    t += .05

    draw()

    #Pause to allow time to draw
    pause(1e-8)

legend()
show()
```
